```python
images = []
scores = []
i = 0
path = "./BreakImages/Folder_0"
for file_name in os.listdir(path):
    image = cv2.imread(f"{path}/{file_name}")
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    image = cv2.resize(image,(200,200))
    image = image.reshape(200*200*3)
    images.append(image)
    scores.append(round(float(file_name[0:4])*100,1))
    if i % 100 == 0:
        print(f"Read {i} images")
    i+=1
```
